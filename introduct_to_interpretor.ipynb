```python
# string can indexed(subscripted)
word = 'python'
word[0] # character in position 0
#word[5] # character in position 5
```

```python
# slice last charecter
# Note: '-0' is same as '0' so last indices start from '-1'
word[-1]
```

```python
# slice second character from last
word[-2]
```

```python
word[-6] # first character from last
```

```python
# indexing: used to extract indidual characters
# slicing : used to extract sub strings
word[0:2] # characters from position 0(include) to 2(exclude)
word[2:5] # characters from position 2 (included) to 5(exclude)
```

```python
# slicing from beging you omit '0' and till required character
word[:2] # character begining from to second character
```

```python
# slicing from position end
word[3:]
```

```python
# start always included and last always excluded
word[:2] + word[2:]
```

```python
word[4:] + word[:4]
```

```python
# Fabonacci series
# sum of two elements defines the next
a, b = 0, 1
while a < 10:
  print(a)
  a, b = b, a+b
```

**Collection or dictionary**

```python
# creat a sample collection
users = { 'Hans': 'active', 'Éléonore': 'inactive', '景太郎': 'active'}
# Strategy: Iterate over copy
for user, status in users.copy().items():
  if status == 'inactive':
    del users[user]
#Strategy: create new collection
active_users = {}
for user, status in users.items():
  if status == 'active':
    active_users[user] = status
```

**Break and Continue**

```python
for n in range(2, 10):
  for x in range(2, n):
    if n%x == 0:
      print(f"{n} equals to {x} {n//x}")
      break
```

```python
for i in range(2,10):
  print(i)
```

```python
for i in range(2, 10):
  if i % 2 == 0:
    print(f"Found an even number {i}")
    continue
  print(f"Found an odd number {i}")
```

**else caluse with loops**

```python
for n in range(2, 10):
  for x in range(2, n):
    if n%x == 0:
      print(n, 'equals', x, '*', n//x)
      break
  else:
    print(n, 'is prime number')
```

**Match Statement**

```python
def http_error(status):
  match status:
    case 400: 'bad request'
    case 404: 'not found'
    case 418: "I'm a tespot"
    case _:
      "something wrong with internet"
```

**Note**: the last block: the “variable name” _ acts as a wildcard and never fails
to match.

You can combine several literals in a single pattern using | (“or”):



```python
case 401 | 403 | 404:
    return "Not allowed"
```
